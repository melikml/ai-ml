```python
# Implementing Logistic Regression from scratch
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def compute_cost(predictions, y):
    eps = 1e-10 
    n = len(y)
    cost = - (1/n) * np.sum(y * np.log(predictions + eps) + (1 - y) * np.log(1 - predictions + eps))
    return cost

# Logistic Regression gradient Descent class
class LogisticRegressionGD:
    def __init__(self, lr=0.01, n_iters=1000, gd_type='batch',
                 batch_size=32, shuffle=True, record_cost=False):
        self.lr = lr
        self.n_iters = n_iters
        self.gd_type = gd_type
        self.batch_size = batch_size
        self.shuffle = shuffle
        self.weights = None
        self.bias = None

        self.record_cost = record_cost
        self.cost_history = [] 

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0.0

        for epoch in range(self.n_iters):
            # Batch Gradient Descent
            if self.gd_type == 'batch':
                linear_pred = np.dot(X, self.weights) + self.bias
                predictions = sigmoid(linear_pred)

                dw = (1/n_samples) * np.dot(X.T, (predictions - y))
                db = (1/n_samples) * np.sum(predictions - y)

                self.weights -= self.lr * dw
                self.bias   -= self.lr * db
            #stochastic gradient descent
            elif self.gd_type == 'stochastic':
                if self.shuffle:
                    indices = np.arange(n_samples)
                    np.random.shuffle(indices)
                else:
                    indices = range(n_samples)

                for idx in indices:
                    xi = X[idx]
                    yi = y[idx]
                    pred = sigmoid(np.dot(xi, self.weights) + self.bias)

                    dw = (pred - yi) * xi
                    db = (pred - yi)

                    self.weights -= self.lr * dw
                    self.bias   -= self.lr * db
            # Mini Batch gradient Descent
            elif self.gd_type == 'mini-batch':
                if self.shuffle:
                    indices = np.arange(n_samples)
                    np.random.shuffle(indices)
                else:
                    indices = range(n_samples)
                for start_idx in range(0, n_samples, self.batch_size):
                    end_idx = start_idx + self.batch_size
                    batch_indices = indices[start_idx:end_idx]

                    X_batch = X[batch_indices]
                    y_batch = y[batch_indices]

                    pred = sigmoid(np.dot(X_batch, self.weights) + self.bias)

                    dw = (1/len(X_batch)) * np.dot(X_batch.T, (pred - y_batch))
                    db = (1/len(X_batch)) * np.sum(pred - y_batch)

                    self.weights -= self.lr * dw
                    self.bias   -= self.lr * db
            if self.record_cost:
                all_preds = sigmoid(np.dot(X, self.weights) + self.bias)
                current_cost = compute_cost(all_preds, y)
                self.cost_history.append(current_cost)

    def predict(self, X):
        linear_pred = np.dot(X, self.weights) + self.bias
        y_pred = sigmoid(linear_pred)
        return (y_pred >= 0.5).astype(int)
```

```python
# Convergence Properties
# The Batch Gradient Descent uses all samples to compute a gradient each epoch.Typically the smoothest convergence curves but each update can be slow for large datasets.
# Stohastic gradient descent updates per sample in random order.It tends to bounce around in the cost surface (you see more noise in the cost curve).It is very good for very large datasets (you get updates immediately rather than summing over all data).
# Mini batch gradient descent splits data into small batches.It is more stable than Stochastic gradient descent alone (less noise in the gradient), yet faster updates than full batch.
```

```python
# Task 4:
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
import torch.nn as nn

class LogisticRegressionModel(nn.Module):
    def __init__(self, n_features):
        super().__init__()
        # A single linear layer: from n_features -> 1
        self.linear = nn.Linear(n_features, 1)

    def forward(self, x):
        # Logistic regression = linear + sigmoid
        return torch.sigmoid(self.linear(x))
n_features = X_train_array.shape[1]
model = LogisticRegressionModel(n_features)
```

```python
def train_model(model, optimizer, X_train, y_train, X_test, y_test, 
                n_epochs=50, batch_size=32):
    criterion = nn.BCELoss()  # Binary cross-entropy for logistic regression
    dataset_size = X_train.shape[0]

    losses = []

    for epoch in range(n_epochs):
        perm = torch.randperm(dataset_size)
        X_train = X_train[perm]
        y_train = y_train[perm]

        for start_idx in range(0, dataset_size, batch_size):
            end_idx = start_idx + batch_size
            xb = X_train[start_idx:end_idx]
            yb = y_train[start_idx:end_idx].view(-1, 1) 

            preds = model(xb)            
            loss  = criterion(preds, yb)

            # backward pass
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
        losses.append(loss.item())

    with torch.no_grad():
        test_preds = model(X_test).view(-1)
        test_preds_class = (test_preds >= 0.5).float()

        y_test_np = y_test.numpy()
        test_preds_np = test_preds_class.numpy()

        acc = accuracy_score(y_test_np, test_preds_np)
        prec = precision_score(y_test_np, test_preds_np)
        rec = recall_score(y_test_np, test_preds_np)
        f1 = f1_score(y_test_np, test_preds_np)

    return {
        'loss_history': losses,
        'accuracy': acc,
        'precision': prec,
        'recall': rec,
        'f1': f1
    }
```

```python
X_test = X_test.to_numpy()
y_test = y_test.to_numpy()
```

```python
X_train_torch = torch.tensor(X_train_array, dtype=torch.float32)
y_train_torch = torch.tensor(y_train_array, dtype=torch.float32)
X_test_torch  = torch.tensor(X_test,  dtype=torch.float32)
y_test_torch  = torch.tensor(y_test,  dtype=torch.float32)
```

```python
import torch.optim as optim

# 1) VANILLA STOCHASTIC GRADIENT DESCENT
model_sgd = LogisticRegressionModel(n_features)
optimizer_sgd = optim.SGD(model_sgd.parameters(), lr=0.01)  
results_sgd = train_model(model_sgd, optimizer_sgd, 
                          X_train_torch, y_train_torch, 
                          X_test_torch, y_test_torch, 
                          n_epochs=50, batch_size=1)

# 2) SGD + MOMENTUM
model_momentum = LogisticRegressionModel(n_features)
optimizer_momentum = optim.SGD(model_momentum.parameters(), lr=0.01, momentum=0.9)
results_mom = train_model(model_momentum, optimizer_momentum,
                          X_train_torch, y_train_torch,
                          X_test_torch, y_test_torch,
                          n_epochs=50, batch_size=1)

# 3) RMSProp
model_rmsprop = LogisticRegressionModel(n_features)
optimizer_rms = optim.RMSprop(model_rmsprop.parameters(), lr=0.001, alpha=0.9)
results_rms = train_model(model_rmsprop, optimizer_rms,
                          X_train_torch, y_train_torch,
                          X_test_torch, y_test_torch,
                          n_epochs=50, batch_size=32)

# 4) ADAM
model_adam = LogisticRegressionModel(n_features)
optimizer_adam = optim.Adam(model_adam.parameters(), lr=0.001, betas=(0.9, 0.999))
results_adam = train_model(model_adam, optimizer_adam,
                           X_train_torch, y_train_torch,
                           X_test_torch, y_test_torch,
                           n_epochs=50, batch_size=32)
```

```python
# Evaluation Metrics
print("Vanilla SGD:", results_sgd)
print("Momentum  :", results_mom)
print("RMSProp   :", results_rms)
print("Adam      :", results_adam)
```

```python
# Hyperparamter Tuning 
possible_lrs = [0.1, 0.01, 0.001]
momentums = [0.0, 0.9]

best_acc = 0.0
best_params = None

for lr_ in possible_lrs:
    for mom_ in momentums:
        # create fresh model
        temp_model = LogisticRegressionModel(n_features)
        temp_optimizer = optim.SGD(temp_model.parameters(), lr=lr_, momentum=mom_)
        res = train_model(temp_model, temp_optimizer,
                          X_train_torch, y_train_torch,
                          X_test_torch, y_test_torch, 
                          n_epochs=50, batch_size=32)
        if res['accuracy'] > best_acc:
            best_acc = res['accuracy']
            best_params = (lr_, mom_)

print(f"Best accuracy: {best_acc}, with lr={best_params[0]} and momentum={best_params[1]}")
```

```python
# Practical Trade Offs:

# From a computational complexity point of view,Momentum, RMSProp, Adam each add a few extra vectors to track (e.g., velocity, squared gradient). This is minor overhead compared to plain SGD.Overall, each iteration is still O(number of parameters)
# Interpretability:The final logistic‐regression model is still interpretable (coefficients). But you must handle an extra layer of complexity in the optimizer’s hyperparameters. For small/medium data sets, all can do well if tuned.
#In deep learning or high‐dim settings, Adam is extremely common. RMSProp also used frequently, especially in RNN tasks. Momentum‐based methods can be equally good.For extremely large data, you’ll likely use mini‐batch with one of these adaptive optimizers.
```
